```python
import re

def hasNumbers(inputString):
    return bool(re.search(r'\d', inputString))


def hasFood(list_criterias):
    if 'Food' not in list_criterias:
        nothing_else = 'Not Specified Food'
        print(nothing_else)
        food = nothing_else
        food_list.append(np.nan)

    
def hasService(list_criterias):
    if 'Service' not in list_criterias:
        nothing_else = 'Not Specified Service'
        print(nothing_else)
        service = nothing_else
        service_list.append(np.nan)
    
def hasValue(list_criterias):
    if 'Value' not in list_criterias:
        nothing_else = 'Not Specified Value'
        print(nothing_else)
        value = nothing_else
        value_list.append(np.nan)
    
def hasAtmosphere(list_criterias):
    if 'Atmosphere' not in list_criterias:
        nothing_else = 'Not Specified Atmosphere'
        print(nothing_else)
        atmosphere = nothing_else
        atmosphere_list.append(np.nan)
    
    
review_list = []
name_list = []
url_list = []
street_list = []
location_list = []
cuisine_list = []
num_of_cuisines_list = []
price_category_list = []
pr_1_list = []
pr_2_list = []
reviewers_list = []
#phone_list = []
ranking_list = []
food_list = []
service_list = []
value_list = []
atmosphere_list = []
count = 1

u2 = u[:]
for url1 in u2:
    url = 'https://www.tripadvisor.com/' + url1
    response = requests.get(url)
    response
    soup = BeautifulSoup(response.text,'html.parser')
    
    #URL
    print(count)
    print(url)
    url_list.append(url)

    
    
    #Name
    all_names = []
    all_names = soup.find_all('li',{"class":"breadcrumb"})
    name = all_names[-1].text
    name_list.append(name)
    print(name)
    
    #Location
    if "Larissa" in url:
        location = "Larissa"
        print("Location: ", location)
        location_list.append(location)
    elif "Trikala" in url:
        location = "Trikala"
        print("Location: ", location)
        location_list.append(location)
    elif "Karditsa" in url:
        location = "Karditsa"
        print("Location: ", location)
        location_list.append(location)
    elif "Volos" in url:
        location = "Volos"
        print("Location: ", location)
        location_list.append(location)
    elif "Magnesia" in url:
        location = "Volos"
        print("Location: ", location)
        location_list.append(location)

        
    
        
    #Street Address
    street = soup.find('a',{"href":"#MAPVIEW"})
    #street = soup.find('span',{"class":"restaurants-detail-overview-cards-LocationOverviewCard__detailLinkText--co3ei"})
    if street ==  None:
        street = 'Not Specified'
        street_list.append(np.nan)
        print(street)
    else:
        street = street.text
        print(street)
        street_list.append(street)
        
    
    #Cuisine type
    #cuisine = soup.find("div",{"class":"restaurants-detail-overview-cards-DetailsSectionOverviewCard__tagText--1OH6h"})
    if soup.find_all('a',{"class":"_2mn01bsa"}):
        cuisine = soup.find_all('a',{"class":"_2mn01bsa"})
        c_list = ''
        i=0
        for c in cuisine:
            temp = c.text
            i = i + 1
            if temp.isalpha() and i != len(cuisine):
                #c_list.append(c.text)
                c_list = c_list + c.text + ', '
            elif temp.isalpha() and i == len(cuisine):
                c_list = c_list + c.text
            

        print(c_list)
        print("Num of cuisines: ", len(cuisine)-1)
        num_of_cuisines = len(cuisine)-1
        num_of_cuisines_list.append(num_of_cuisines)
        cuisine_list.append(c_list)
    else:
        cuisine = 'Not Specified'
        num_of_cuisines = 'Not Specified'
        num_of_cuisines_list.append(np.nan)
        cuisine_list.append(np.nan)
        print(cuisine)
        print(num_of_cuisines)
        
    
    
    #price category
    price_category = soup.find('div',{"class":"_1XLfiSsv"})
    if  price_category == None:
        price_category = 'Not Specified a'
        pr_1 = 'Not Specified price_1'
        pr_2 = 'Not Specified price_2'
        print(pr_1)
        print(pr_2)
        print(price_category)
        price_category_list.append(np.nan)
        pr_1_list.append(np.nan)
        pr_2_list.append(np.nan)
    else:
        price_category = price_category.text
        has_or_not = hasNumbers(price_category)
        if has_or_not == False:
            price_category = 'Not Specified b'
            pr_1 = 'Not Specified price_1'
            pr_2 = 'Not Specified price_2'
            print(pr_1)
            print(pr_2)
            print(price_category)
            price_category_list.append(np.nan)
            pr_1_list.append(np.nan)
            pr_2_list.append(np.nan)
        else:
            print(price_category)
            pr_1 = ''
            pr_2 = ''
            for i in price_category:
                if i == ' ':
                    break
                if i.isdigit():
                    pr_1 = pr_1 + i
            
            for c in reversed(price_category):
                if c == '€':
                    break
                if c != '€':
                    pr_2 = pr_2 + c
            
            pr_2 = pr_2[::-1]
            pr_1 = int(pr_1)
            pr_2 = int(pr_2)

            print("Price_1: ", pr_1)
            print("Price_2: ", pr_2)
            
            pr_1_list.append(pr_1)
            pr_2_list.append(pr_2)
            price_category_list.append(price_category)
    
    
    #location-based ranking
    if soup.find('a',{"class":"_15QfMZ2L"}):
        if soup.find('a',{"class":"_15QfMZ2L"}).find('span').find('b'):
            ranking = soup.find('a',{"class":"_15QfMZ2L"}).find('span').find('b')
            ranking = ranking.text
            ranking = ranking[1:]
            ranking_list.append(ranking)
            print("Ranking: ",ranking)
        else:
            ranking = 'Not Specified'
            ranking_list.append(np.nan)
            print(ranking)        
    else:
        ranking = 'Not Specified'
        ranking_list.append(np.nan)
        print(ranking)
        
    
    
    #ratings of the four criteria
    if soup.find_all('span', {"class":"_2vS3p6SS"}):
        criterias = soup.find_all('span', {"class":"_2vS3p6SS"})
        if  len(criterias) == 0:
            criterias = 'Not Specified'
            print(criterias)
            food = criterias
            service = criterias
            value = criterias
            atmosphere = criterias
            food_list.append(np.nan)
            service_list.append(np.nan)
            value_list.append(np.nan)
            atmosphere_list.append(np.nan)
        else:
            num_criterias = len(criterias)
            
            list_criterias = []
            for n in criterias:
                if n.text == 'Food':
                    list_criterias.append(n.text)
                elif n.text == 'Service':
                    list_criterias.append(n.text)
                elif n.text == 'Value':
                    list_criterias.append(n.text)
                elif n.text == 'Atmosphere':
                    list_criterias.append(n.text)
                    
            if num_criterias == 1:
                hasFood(list_criterias)
                hasService(list_criterias)
                hasValue(list_criterias)
                hasAtmosphere(list_criterias)
                

            elif num_criterias == 2:
                hasFood(list_criterias)
                hasService(list_criterias)
                hasValue(list_criterias)
                hasAtmosphere(list_criterias)
                

            elif num_criterias == 3:
                hasFood(list_criterias)
                hasService(list_criterias)
                hasValue(list_criterias)
                hasAtmosphere(list_criterias)
                


            num_criterias = num_criterias + 1
            ratings = soup.find_all('span', {'class': re.compile('ui_bubble_rating bubble_\d*')})
            ratings = re.findall('\d+', ''.join(map(str, ratings)))
            ratings_selection = ratings[1:num_criterias]
            int_list = []

            for rating in ratings_selection:
                int_list.append(int(rating))

            k=0
            for criteria in criterias:
                temp = int_list[k]/10
                print(criteria.text + '\t' + str(temp))
                k = k + 1
                if criteria.text == 'Food':
                    food = temp
                    food_list.append(food)
                elif criteria.text == 'Service':
                    service = temp
                    service_list.append(service)
                elif criteria.text == 'Value':
                    value = temp
                    value_list.append(value)
                elif criteria.text == 'Atmosphere':
                    atmosphere = temp
                    atmosphere_list.append(atmosphere)
    else:
        criterias = 'Not Specified criterias'
        print(criterias)
        food = criterias
        service = criterias
        value = criterias
        atmosphere = criterias
        food_list.append(np.nan)
        service_list.append(np.nan)
        value_list.append(np.nan)
        atmosphere_list.append(np.nan)
    
    
    
    
    
    #Review ratings overview
    #l = soup.find('span',{"class":"restaurants-detail-overview-cards-RatingsOverviewCard__overallRating--nohTl"})
    if soup.find('span',{"class":"r2Cf69qf"}):
        review = soup.find('span',{"class":"r2Cf69qf"})
        r = review.text
        review_list.append(r)
        print(r)
    else:
        review = "No review"
        review_list.append(np.nan)
        print(review)
        
        
        
    
    #No of Reviewers 
    reviewers = soup.find('span',{"class":"_3Wub8auF"})
    if reviewers == None:
        reviewers = 'Undefined Number'
        reviewers_list.append(np.nan)
        print(reviewers)
        
    else:
        reviewers = reviewers.text 
        reviewers_list.append(reviewers)
        print(reviewers)
    
    
        
        
    count +=1   
    print('')
     
    #diction = {'Name': name_list,'Cuisine': cuisine_list,'Location': locality_list,'Price Category': price_category_list,'Reviews' : review_list,'No of Reviews' : reviewers_list,'Contact Number': phone_list,'Trip_advisor Url': url_list,'Food' : food_list,'Service' : service_list,'Value' : value_list,'Atmosphere' : atmosphere_list,'Ranking': ranking_list}
    #df = pd.DataFrame.from_dict(diction, orient='index')
    df = pd.DataFrame({'Name': name_list,
                       'Cuisine': cuisine_list,
                       'Street': street_list,
                       'Location': location_list,
                    'Price Category': price_category_list,
                    'Price_1': pr_1_list,
                    'Price_2': pr_2_list,
                    'Total Rating' : review_list,
                    'Num of Reviews' : reviewers_list,
                    'Num of Cuisines': num_of_cuisines_list,
                    'Trip_advisor Url': url_list,
                    'Food' : food_list,
                    'Service' : service_list,
                    'Value' : value_list,
                    'Atmosphere' : atmosphere_list,
                    'Ranking': ranking_list
                    })
    
    
    df = df.append({'Name': name, 
                       'Cuisine': cuisine,
                       'Street': street,
                        'Location': location,
                        'Price Category': price_category,
                        'Price_1': pr_1,
                        'Price_2': pr_2,
                        'Total Rating' : review,
                        'Num of Reviews' : reviewers,
                        'Num of Cuisines': num_of_cuisines,
                        'Trip_advisor Url': url,
                        'Food' : food,
                        'Service' : service,
                        'Value' : value,
                        'Atmosphere' : atmosphere,
                        'Ranking': ranking
                   },ignore_index=True)
    
    #print(aelara)
```
